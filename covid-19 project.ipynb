```python
X = df.drop(["Country/Region","New cases","New deaths","New recovered","Deaths / 100 Cases","WHO Region","1 week % increase","1 week change","Recovered / 100 Cases","Confirmed last week","Deaths / 100 Recovered"],axis=1)
Y = df["Confirmed last week"]
```

```python
X.head()
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,Y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
coeff_df = pd.DataFrame(regression.coef_,X.columns,columns=["Coefficient"])
coeff_df
```

```python
Y_pred = regression.predict(X_train)
print(Y_pred)
```

```python
df = pd.DataFrame({'Actual' : Y_train, 'Predicted' : Y_pred})
df.tail()
```

```python
from sklearn import metrics
print("Mean Absolute Error :",metrics.mean_absolute_error(Y_train,Y_pred))
print("Mean Square Error :",metrics.mean_absolute_error(Y_train,Y_pred))
print("Root Mean Squared Error :",np.sqrt(metrics.mean_squared_error(Y_train,Y_pred)))
```

```python
from sklearn.metrics import r2_score
print("R2 error :",metrics.r2_score(Y_train,Y_pred))
```
